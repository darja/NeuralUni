# Классическое машинное обучение: признаки для разных типов данных

Вычислить HoG признаки для Lena30.jpg. Можно с помощью opencv или skimage.feature. Визуализировать результат.

# Подготовка окружения

```
import numpy as np
import matplotlib.pyplot as plt

from skimage.feature import hog
from skimage import data, color, exposure
```

# Чтение файла

```
rgbImage = data.imread('lena30.jpg')
gsImage = color.rgb2gray(rgbImage)

# Демонстрация файлов
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
  
ax1.axis('off')
ax1.imshow(rgbImage, cmap=plt.cm.gray)
ax1.set_title('Original')
  
ax2.axis('off')
ax2.imshow(gsImage, cmap=plt.cm.gray)
ax2.set_title('Grayscale')
plt.show()
```

# Вычисление HOG-дескрипторов

```
fd, hog_image = hog(gsImage, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1), block_norm="L1", visualize=True)
```

# Визуализация

## Гистограмма

```
plt.hist(fd)
plt.show()
```

## Визуализация градиентов

```
hog_image_highlighted = exposure.rescale_intensity(hog_image, in_range=(0, 0.02))
plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(hog_image_highlighted)
```
